**1. Research about the Spectral Clustering method, and answer the following questions:**

Spectral clustering is a technique derived from graph theory, used to identify communities of nodes in a graph based on the edges connecting them. The method is flexible and allows for clustering non-graph data as well.

It is primarily characterized by being non-convex, meaning it lacks surfaces that exhibit curvature.

**a. ¿In which cases might it be more useful to apply?**


- **High-dimensional data:** In the realm of high-dimensional data where visualizing and understanding underlying patterns can be challenging, Spectral Clustering shines by providing dimensionality reduction capabilities. By transforming high-dimensional data into a lower-dimensional space, Spectral Clustering facilitates easier visualization and pattern recognition.

- **Graph-structured data:**  In scenarios where the data exhibits a graph-like structure, such as social networks or interconnected systems, Spectral Clustering emerges as a promising choice for effectively grouping nodes into meaningful communities.

- **Data with clusters of different sizes:** Spectral Clustering showcases its versatility by effectively managing clusters of diverse sizes. Its lack of assumptions regarding cluster shape allows it to accommodate clusters of different magnitudes, offering robust performance across a wide range of datasets.

- **Anomaly detection:**: Spectral Clustering proves valuable in anomaly detection tasks by leveraging its foundation in data similarity. By identifying deviations from the norm based on data point relationships, Spectral Clustering aids in detecting outliers or anomalous instances within the dataset.

- **Nonlinear data:** Spectral Clustering demonstrates its superiority over traditional methods like K-means when handling datasets with nonlinear structures. Unlike K-means, which assumes linearity, Spectral Clustering adapts well to nonlinear relationships within the data, enhancing clustering accuracy.




**b. ¿What are the mathematical fundamentals of it?**

The mathematical fundamentals of Spectral Clustering involve several key concepts:

- **Graph Laplacian:** At the core of Spectral Clustering lies the graph Laplacian matrix. This matrix is derived from the affinity matrix, which quantifies the similarity between data points. The graph Laplacian matrix captures the graph's structure and plays a crucial role in spectral clustering algorithms.

- **Eigenvalue Decomposition:** Spectral Clustering involves computing the eigenvectors and eigenvalues of the graph Laplacian matrix. Eigenvectors represent directions within the data space, while eigenvalues indicate the variance along these directions. By decomposing the Laplacian matrix, Spectral Clustering extracts meaningful information about the data's structure.

- **Graph Partitioning:** Spectral Clustering utilizes spectral embedding to project data onto a lower-dimensional space defined by eigenvectors. This step facilitates clustering by transforming the data into a representation conducive to partitioning.

- **Cluster Assignment:** Following spectral embedding, Spectral Clustering often employs clustering algorithms like k-means to partition data into distinct clusters. Alternatively, it can directly use spectral properties for clustering without additional algorithms.

- **Normalization and Regularization:** Techniques such as symmetric normalization and regularization are applied during Laplacian matrix computation to enhance performance and ensure robustness across different datasets.

**c. ¿What is the algorithm to compute it?**

- **Affinity Matrix:** Initially, the affinity matrix is constructed, which measures the similarity between data points. This matrix can be built in different ways, for example, using a distance function or a similarity function.

- **Laplacian Graph:** Using the affinity matrix, compute the graph Laplacian matrix. There are different types of Laplacians, including unnormalized, normalized (such as symmetric or random walk normalized), and regularized versions. The choice depends on the specific application and desired properties.

- **Eigenvalue decomposition:** OOnce the affinity matrix is obtained, eigenvalue decomposition is used to find the eigenvectors and eigenvalues of the matrix. The eigenvectors corresponding to the largest eigenvalues are used to construct a new representation of the data in a lower-dimensional space.

- **Clustering:** Finally, a clustering algorithm, such as k-means, is applied to the spectral embeddings obtained in the previous step. Alternatively, spectral clustering can directly partition data based on spectral properties without the need for additional clustering algorithms.


**d. ¿Does it hold any relation to some of the concepts previously mentioned in class? Which, and how?**

- **Eigenvalues and Eigenvectors:** Spectral clustering leverages eigenvalues and eigenvectors of an affinity matrix to reduce data dimensionality and conduct clustering in a lower-dimensional space. Eigenvectors serve as new data representations, with eigenvalues guiding the selection of the most significant eigenvectors.

- **Dimensionality Reduction:** Following dimensionality reduction, clustering algorithms like K-means are applied to the reduced-dimensional space to cluster the data. This reduction often enhances clustering efficiency and decreases computational complexity, particularly with large datasets.

- **K-means Algorithm:** Spectral clustering and the K-means algorithm are both clustering techniques known for their non-convex nature, meaning they don't exhibit curved surfaces. They are utilized to group a dataset into clusters. While they share similarities, they differ in their methodologies and applications. Spectral clustering may employ the K-means algorithm to cluster data in a lower-dimensional space. However, the transformation of data into this space via eigenvalue decomposition distinguishes spectral clustering from the partitioning approach of K-means.



**2. Research about the DBSCAN method, and answer the following questions:**


This algorithm serves as a foundation for density-based clustering. It's adept at identifying clusters of various shapes and sizes within vast datasets, even amidst noise and outliers.

**a. ¿In which cases might it be more useful to apply?**

- DBSCAN is particularly useful when dealing with datasets that contain clusters of varying shapes and sizes.

- It is effective in identifying clusters in data with irregular shapes or densities.

- DBSCAN performs well in the presence of noise and outliers since it does not assume any particular distribution of data points.
- It is suitable for datasets where traditional distance-based clustering algorithms struggle, such as datasets with varying densities.


**b. ¿What are the mathematical fundamentals of it?**

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is based on the concept of density reachability.

It defines two parameters: epsilon (ε) and minPts:
- epsilon (ε) defines the radius within which neighboring points are considered part of the same cluster.

- minPts specifies the minimum number of points required within the epsilon radius to form a dense region (core point).

The algorithm starts by randomly selecting a point and finding its neighbors within the epsilon radius.

If the number of neighbors is greater than or equal to minPts, the point is labeled as a core point and all reachable points from this core point form a cluster.
Points that do not meet the criteria to be core points or reachable from core points are labeled as noise or outliers.

**c. ¿Is there any relation between DBSCAN and Spectral Clustering? If so, what is it?**

DBSCAN and Spectral Clustering are clustering algorithms with distinct strengths and weaknesses, suitable for various scenarios. It's feasible to employ both algorithms in combination, leveraging DBSCAN to identify clusters in a dataset and then refining the clustering results using Spectral Clustering.

For instance, DBSCAN may act as a preprocessing step to detect outliers or noisy data before applying Spectral Clustering. Furthermore, hybrid clustering approaches can benefit from both algorithms to enhance clustering accuracy.

**3. ¿What is the elbow method in clustering? And which flaws does it pose to assess quality?**


The elbow method in clustering is a technique used to determine the optimal number of clusters in a dataset. It involves plotting a clustering metric, such as within-cluster sum of squares (WCSS), against the number of clusters and identifying the point where the rate of decrease in the metric slows down, resembling an "elbow" shape. This point is often interpreted as the optimal number of clusters.

However, the elbow method for clustering has limitations:

Subjectivity: Choosing the "elbow" point is subjective.

Single Metric Reliance: It relies on a single metric, potentially missing nuances in the data.

Shape Insensitivity: Not effective for non-convex cluster shapes.

Context Oversight: Doesn't consider specific data characteristics or context.

While it provides a basic estimate of cluster number, it's best used cautiously alongside other evaluation methods.

**c.** For both k-means and k-medoids (your implementations), calculate the silhouette plots and coefficients for each run, iterating K from 1 to 5 clusters.

````python
for n_clusters in range(2,6):
    fig, axs = plt.subplots(1, 2, figsize=(20, 5))
    K_medoids2=K_Medoids()
    labels=K_medoids2.fit_transform(X=X,K=n_clusters,max_iters=1000)
    silhouette_vals = silhouette_samples(X, labels, metric='euclidean')
    silhouette_avg = silhouette_score(X, labels, metric='euclidean')
    y_lower, y_upper = 0, 0
    yticks = []
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        axs[0].barh(range(y_lower, y_upper), cluster_silhouette_vals, height=1.0,
                edgecolor='none', color=color)
        yticks.append((y_lower + y_upper) / 2.)
        y_lower += len(cluster_silhouette_vals)
    axs[0].axvline(silhouette_avg, color="red", linestyle="--")
    axs[0].set_yticks(yticks,np.unique(labels))
    axs[0].set_ylabel('Cluster')
    axs[0].set_xlabel('Silhouette coefficient')
    axs[0].set_title('Medoids')

    K_means=K_Means()
    labels=K_means.fit_transform(X=X,K=n_clusters,max_iters=1000)
    silhouette_vals = silhouette_samples(X, labels, metric='euclidean')
    silhouette_avg = silhouette_score(X, labels, metric='euclidean')
    y_lower, y_upper = 0, 0
    yticks = []
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        color = cm.jet(float(i) / n_clusters)
        axs[1].barh(range(y_lower, y_upper), cluster_silhouette_vals, height=1.0,
                edgecolor='none', color=color)
        yticks.append((y_lower + y_upper) / 2.)
        y_lower += len(cluster_silhouette_vals)
    axs[1].axvline(silhouette_avg, color="red", linestyle="--")
    axs[1].set_yticks(yticks,np.unique(labels))
    axs[1].set_ylabel('Cluster')
    axs[1].set_xlabel('Silhouette coefficient')
    axs[1].set_title('Means')
````
